```python
df3=pd.read_csv("A_Z_medicines_dataset_of_India.csv")
df3.head()
```

```python
print(df3.describe())  # Display summary statistics of numeric columns
```

```python
type_price_summary = df3.groupby('type')['price(₹)'].mean().sort_values(ascending=False)
```

```python
# Plotting the database on the basis of the type of medicine and its price 
# So that we can predict which medicine has the highest price

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(x=type_price_summary.index, y=type_price_summary.values)
plt.title('Average Price of Medicines by Type')
plt.xlabel('Medicine Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# From this plot we can predict the sale of medicines on daily basis which are consumed more ,so that we can keep the 
# stock of those medicines

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df4=pd.read_csv("salesdaily.csv")
df4.head(10)



df4['datum'] = pd.to_datetime(df4['datum'])


plt.figure(figsize=(12, 6))
plt.plot(df4['datum'], df4['M01AB'], label='Medicine A')
plt.plot(df4['datum'], df4['M01AE'], label='Medicine B')
plt.plot(df4['datum'], df4['N02BA'], label='Medicine C')
plt.plot(df4['datum'], df4['N02BE'], label='Medicine D')
plt.plot(df4['datum'],df4['N05C'], label='Medicine E')
plt.plot(df4['datum'],df4['R03'], label='Medicine F')

plt.plot(df4['datum'],df4['R06'], label='Medicine G')


# Adding labels and legend
plt.title('Medicine Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
